Data Cleaning

```python
#converting date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
#creating additional columns
df['Year']=df['Order Date'].dt.year
df['Month']=df['Order Date'].dt.month
df['Month Name']=df['Order Date'].dt.month_name
df['Quarter'] = df['Order Date'].dt.quarter
df['Day of Week'] = df['Order Date'].dt.day_name()
```

```python
#calculating shipping time
df['Shipping Days']=(df['Ship Date']-df['Order Date']).dt.days
```

Key Business Metrics

```python
total_sales=df['Sales'].sum()
total_profit=df['Profit'].sum()
total_orders=df['Order ID'].nunique()
total_customers=df['Customer ID'].nunique()
total_products=df['Product ID'].nunique()
avg_shipping_time=df['Shipping Days'].mean()
avg_order_value = df.groupby('Order ID')['Sales'].sum().mean()
profit_margin = (total_profit / total_sales) * 100
```

```python
print("KEY BUSINESS METRICS")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
print(f"Total Orders: {total_orders:,}")
print(f"Total Customers: {total_customers:,}")
print(f"Total Products: {total_products:,}")
print(f"Average Shipping Time: {avg_shipping_time:.2f} days")

print(f"Average Order Value: ${avg_order_value:.2f}")
print(f"Orders per Customer: {total_orders/total_customers:.2f}")
```

Category Analysis

```python
# Category Analysis

print("REVENUE ANALYSIS BY CATEGORY")
print("\n")

category_analysis = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique',
    'Quantity': 'sum'
}).round(2)

category_analysis['Profit Margin %'] = (category_analysis['Profit'] / category_analysis['Sales'] * 100).round(2)
category_analysis['Avg Sale per Order'] = (category_analysis['Sales'] / category_analysis['Order ID']).round(2)
category_analysis = category_analysis.sort_values('Sales', ascending=False)

# Renaming columns for clarity
category_analysis.columns = ['Total Sales ($)', 'Total Profit ($)', 'Number of Orders',
                              'Units Sold', 'Profit Margin (%)', 'Avg Sale/Order ($)']

print(category_analysis)
```

```python
# Visualization: Sales by Category
plt.figure(figsize=(10, 6))
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
colors = ['#2ecc71', '#3498db', '#e74c3c']
category_sales.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Total Sales by Category', fontsize=16, fontweight='bold')
plt.ylabel('Sales ($)', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)


for i, v in enumerate(category_sales.values):
    plt.text(i, v + 1000, f'${v:,.0f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# VISUALIZATIONS
plt.figure(figsize=(10, 6))
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Category Performance Analysis', fontsize=18, fontweight='bold')

# 1. Sales by Category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
colors = ['#2ecc71', '#3498db', '#e74c3c']
axes[0, 0].bar(category_sales.index, category_sales.values, color=colors, edgecolor='black')
axes[0, 0].set_title('Total Sales by Category', fontsize=14, fontweight='bold')
axes[0, 0].set_ylabel('Sales ($)', fontsize=12)
axes[0, 0].grid(axis='y', alpha=0.3)
for i, v in enumerate(category_sales.values):
    axes[0, 0].text(i, v + 10000, f'${v/1000:.0f}K', ha='center', fontweight='bold', fontsize=10)

# 2. Profit by Category
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
axes[0, 1].bar(category_profit.index, category_profit.values, color=colors, edgecolor='black')
axes[0, 1].set_title('Total Profit by Category', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Profit ($)', fontsize=12)
axes[0, 1].grid(axis='y', alpha=0.3)
for i, v in enumerate(category_profit.values):
    axes[0, 1].text(i, v + 2000, f'${v/1000:.0f}K', ha='center', fontweight='bold', fontsize=10)

# 3. Profit Margin by Category
category_margin = (df.groupby('Category')['Profit'].sum() / df.groupby('Category')['Sales'].sum() * 100).sort_values(ascending=False)
axes[1, 0].bar(category_margin.index, category_margin.values, color=['#f39c12', '#9b59b6', '#1abc9c'], edgecolor='black')
axes[1, 0].set_title('Profit Margin by Category', fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel('Profit Margin (%)', fontsize=12)
axes[1, 0].grid(axis='y', alpha=0.3)
axes[1, 0].axhline(y=12.47, color='red', linestyle='--', linewidth=2, label='Overall Avg (12.47%)')
for i, v in enumerate(category_margin.values):
    axes[1, 0].text(i, v + 0.5, f'{v:.1f}%', ha='center', fontweight='bold', fontsize=10)
axes[1, 0].legend()

# 4. Orders Distribution
category_orders = df.groupby('Category')['Order ID'].nunique().sort_values(ascending=False)
axes[1, 1].pie(category_orders.values, labels=category_orders.index, autopct='%1.1f%%',
               colors=colors, startangle=90, explode=(0.05, 0, 0))
axes[1, 1].set_title('Order Distribution by Category', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nKey Insights:")
print(f"• Highest revenue category: {category_sales.idxmax()} (${category_sales.max():,.2f})")
print(f"• Most profitable category: {category_profit.idxmax()} (${category_profit.max():,.2f})")
print(f"• Best profit margin: {category_margin.idxmax()} ({category_margin.max():.2f}%)")
```

```python
# SUB-CATEGORY ANALYSIS

subcategory_analysis = df.groupby('Sub-Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'
}).round(2)

subcategory_analysis['Profit Margin %'] = (subcategory_analysis['Profit'] / subcategory_analysis['Sales'] * 100).round(2)
subcategory_analysis = subcategory_analysis.sort_values('Sales', ascending=False)

print("SUB-CATEGORIES BY SALES\n")
print(subcategory_analysis)
```

```python
# VISUALIZATIONS
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1. Top Sub-Categories by Sales
top10_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
axes[0].barh(top10_sales.index, top10_sales.values, color='skyblue', edgecolor='black')
axes[0].set_title('Top Sub-Categories by Sales', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Sales ($)', fontsize=12)
axes[0].invert_yaxis()
axes[0].grid(axis='x', alpha=0.3)
for i, v in enumerate(top10_sales.values):
    axes[0].text(v + 2000, i, f'${v/1000:.0f}K', va='center', fontweight='bold')

# 2. Top Sub-Categories by Profit
top10_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
axes[1].barh(top10_profit.index, top10_profit.values, color='lightgreen', edgecolor='black')
axes[1].set_title('Top Sub-Categories by Profit', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Profit ($)', fontsize=12)
axes[1].invert_yaxis()
axes[1].grid(axis='x', alpha=0.3)
for i, v in enumerate(top10_profit.values):
    axes[1].text(v + 500, i, f'${v/1000:.0f}K', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

Discount Impact Analysis

```python
df['Discount Bin'] = pd.cut(df['Discount'], bins=[-0.01, 0, 0.1, 0.2, 0.3, 1.0],
                              labels=['No Discount', '1-10%', '11-20%', '21-30%', '30%+'])

discount_analysis = df.groupby('Discount Bin').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'
}).round(2)

discount_analysis['Profit Margin %'] = (discount_analysis['Profit'] / discount_analysis['Sales'] * 100).round(2)
discount_analysis['Avg Order Value'] = (discount_analysis['Sales'] / discount_analysis['Order ID']).round(2)

print(discount_analysis)
```

```python
# Visualization: Discount Impact
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Discount Impact Analysis', fontsize=18, fontweight='bold')

# Profit Margin by Discount Level
discount_margin = (df.groupby('Discount Bin')['Profit'].sum() / df.groupby('Discount Bin')['Sales'].sum() * 100)
axes[0].bar(discount_margin.index, discount_margin.values,
            color=['#2ecc71', '#f39c12', '#e67e22', '#e74c3c', '#c0392b'], edgecolor='black')
axes[0].set_title('Profit Margin by Discount Level', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Profit Margin (%)', fontsize=12)
axes[0].set_xlabel('Discount Range', fontsize=12)
axes[0].axhline(y=12.47, color='blue', linestyle='--', linewidth=2, label='Overall Avg (12.47%)')
axes[0].grid(axis='y', alpha=0.3)
axes[0].legend()
axes[0].tick_params(axis='x', rotation=45)
for i, v in enumerate(discount_margin.values):
    axes[0].text(i, v + 0.5, f'{v:.1f}%', ha='center', fontweight='bold')

# Sales vs Profit by Discount
discount_sales = df.groupby('Discount Bin')['Sales'].sum() / 1000
discount_profit = df.groupby('Discount Bin')['Profit'].sum() / 1000
x = range(len(discount_sales))
width = 0.35
axes[1].bar([i - width/2 for i in x], discount_sales.values, width, label='Sales ($K)',
            color='skyblue', edgecolor='black')
axes[1].bar([i + width/2 for i in x], discount_profit.values, width, label='Profit ($K)',
            color='lightgreen', edgecolor='black')
axes[1].set_title('Sales vs Profit by Discount Level', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Amount ($K)', fontsize=12)
axes[1].set_xlabel('Discount Range', fontsize=12)
axes[1].set_xticks(x)
axes[1].set_xticklabels(discount_sales.index, rotation=45)
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
print("\n OVERALL BUSINESS PERFORMANCE 📊\n")

print(f"Total Revenue:              ${df['Sales'].sum():>15,.2f}")
print(f"Total Profit:               ${df['Profit'].sum():>15,.2f}")
print(f"Overall Profit Margin:      {(df['Profit'].sum()/df['Sales'].sum()*100):>15.2f}%")
print(f"Total Orders:               {df['Order ID'].nunique():>15,}")
print(f"Total Customers:            {df['Customer ID'].nunique():>15,}")
print(f"Average Order Value:        ${df.groupby('Order ID')['Sales'].sum().mean():>15,.2f}")
print(f"Orders per Customer:        {df['Order ID'].nunique()/df['Customer ID'].nunique():>15.2f}")

print("\n TOP PERFORMERS\n")
cat_sales = df.groupby('Category')['Sales'].sum()
cat_profit = df.groupby('Category')['Profit'].sum()
print(f"Best Category (Sales):      {cat_sales.idxmax():>20} (${cat_sales.max():,.2f})")
print(f"Best Category (Profit):     {cat_profit.idxmax():>20} (${cat_profit.max():,.2f})")

region_profit = df.groupby('Region')['Profit'].sum()
print(f"Best Region (Profit):       {region_profit.idxmax():>20} (${region_profit.max():,.2f})")

subcat_profit = df.groupby('Sub-Category')['Profit'].sum()
print(f"Best Sub-Category:          {subcat_profit.idxmax():>20} (${subcat_profit.max():,.2f})")

segment_sales = df.groupby('Segment')['Sales'].sum()
print(f"Best Segment:               {segment_sales.idxmax():>20} (${segment_sales.max():,.2f})")

print("\n PROBLEM AREAS\n")
cat_margin = (df.groupby('Category')['Profit'].sum() / df.groupby('Category')['Sales'].sum() * 100)
print(f"Lowest Margin Category:     {cat_margin.idxmin():>20} ({cat_margin.min():.2f}%)")

region_margin = (df.groupby('Region')['Profit'].sum() / df.groupby('Region')['Sales'].sum() * 100)
print(f"Lowest Margin Region:       {region_margin.idxmin():>20} ({region_margin.min():.2f}%)")

unprofitable_subs = df.groupby('Sub-Category')['Profit'].sum()
unprofitable_subs = unprofitable_subs[unprofitable_subs < 0]
if len(unprofitable_subs) > 0:
    print(f"Unprofitable Sub-Categories: {len(unprofitable_subs):>19} (Losing ${abs(unprofitable_subs.sum()):,.2f})")

print("\n SEASONALITY \n")

monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
print(f"Best Month:                 {str(monthly_sales.idxmax()):>20} (${monthly_sales.max():,.2f})")
print(f"Worst Month:                {str(monthly_sales.idxmin()):>20} (${monthly_sales.min():,.2f})")
print(f"Sales Volatility:           {(monthly_sales.std()/monthly_sales.mean()*100):>19.1f}% (variation)")

print("\n CUSTOMER INSIGHTS \n")
top_customer = df.groupby('Customer Name')['Sales'].sum().idxmax()
top_customer_sales = df.groupby('Customer Name')['Sales'].sum().max()
print(f"Top Customer:               {top_customer[:25]:>20}")
print(f"Top Customer Value:         ${top_customer_sales:>15,.2f}")
print(f"Top 10 Customers:           ${df.groupby('Customer Name')['Sales'].sum().nlargest(10).sum():>15,.2f} ({df.groupby('Customer Name')['Sales'].sum().nlargest(10).sum()/df['Sales'].sum()*100:.1f}% of revenue)")
```
